<br>
<h3>Data Cleaning & Data Manipulate</h3>
<br>

```python
# Rename the 'Country/Region' column to 'Date'
df = df.rename(columns = {'Country/Region':'Date'})

# Drop the first row of the dataframe and reset the index
df = df.drop([0],axis = 0).reset_index(drop = True)

# Create a new column 'World' which is the sum of all columns except for the 'Date' column
df['World'] = df.iloc[:,1:].sum(axis = 1)

# Convert the 'Date' column from a string to a datetime object
df['Date'] = pd.to_datetime(df['Date'],format = '%m/%d/%y')

# Convert the 'Date' column from a datetime object to just a date
df['Date'] = df['Date'].dt.date

df1 = df.copy()
df1[-5:].T
```

<br>
<h3>Average by country</h3>
<br>

```python
# Get a list of all the column names
col = df.columns.tolist()

# Loop over the column names, starting from the second column
for i in range(1,len(col)):
    item = col[i]
    df1[item] = df[item].rolling(window = 7).mean()
    
# Show the last 7 rows   
df1[-7:]
```

<br>
<h3>Weekly average COVID infection rate in each country</h3>
<br>

```python
# Create a new dataframe 'df2'
df2 = df1.copy()

# Loop over the rows of the dataframe, starting from the 7th row
for i in range (7,len(df2)):
    
    # Loop over the columns of the dataframe, starting from the second column
    for j in range (1,len(col)):
        
        # Compute the 7-day average
        df2.iloc[i,j] = (df1.iloc[i,j] - df1.iloc[i-7,j]) / 7

# Store the last value in the 'Date' column        
df = df2.iloc[-1,0]

# Show the last 7 rows
df2[-7:]
```

<br>
<h3>Check out the information after New Year's Eve</h3>
<br>

```python
# Create a new dataframe 'df3' contains the last 28 rows
df3 = df2.tail(28)

# create a new figure and axis object
fig, ax = plt.subplots(figsize=(7,5))

# create a twin axis object
ax2 = ax.twinx()

# Plot 'World' against 'Date' column on the first axis
ax.plot(df3['Date'], df3['World'], label='World')

# Plot 'World' against 'Date' column on the second axis
ax2.plot(df3['Date'], df3['World'], alpha=0)

# Set the x-axis tick labels
ax.tick_params(axis='x', labelrotation=90)

# set the title
ax.set_title('World weekly mean increase')

# set the label
ax.set_xlabel('Date')
ax.set_ylabel('Weekly mean increase')
ax2.set_ylabel('Weekly mean increase')

ax.legend()
plt.show()
```

<br>
<h3>Create a table to compare</h3>
<br>

```python
# Drop the 'World' column
df3 = df3.drop('World',axis =1)

# Create a new dataframe 'df4' selecting the last row (excluding the 'Date' column)
# Sort the values
# Rename the resulting
# Reset the index
# Rename the column

df4 = (
        df3.iloc[-1, 1:]
        .sort_values(ascending=False)
        .rename("weekly_mean_increase")
        .reset_index()     
        .rename(columns={'index': 'country'})
      )


df4
```

<br>
<h3>Explore each country where COVID is spreading</h3>
<br>

```python
# Slice the first 10 rows
df5 = df4[:10]

# Plot a bar chart
plt.bar(df5['country'], df5['weekly_mean_increase'])

# Set the label
plt.xlabel('country')
plt.ylabel('Weekly mean increase')

# Set the title
plt.title('The lastest ranking of weekly mean increase ('+str(df)+')')

# Rotate the x-axis labels
plt.xticks(rotation = 90)

plt.show()
```
